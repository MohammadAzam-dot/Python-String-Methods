```python
# Solution 2 - (Diff between join() and +operator)

# (+) Operator directly combines two or more strings
# Simple and easy for small number of strings

string1 = "Hello"
string2 = "World"

result = string1 + ' ' + string2
print(result)


# join()

# Used to combine multiple strings from an iterable (list,tuple,etc)
# More efficient and faster when joining multiple strings 
# uses a separator string

words = ["Hello","Python","World"]
result = " ".join(words)
print(result)
```

```python
# Solution 19

# We can use slicing to get a part (substring) of a string.
# Syntax: string[start:stop:step]

text = "hello python"
print(text[0:5])   # Output: hello
print(text[6:])    # Output: python
print(text[:5])    # Output: hello
print(text[::])    # Ouput: hello python
```

```python
# Solution 20

# We can remove specific characters from a string using the replace() method 
# or the maketrans() and translate() methods.

# Using replace()
text = "hello-python-world"
print(text.replace("-", ""))   # Output: hellopythonworld

# remove spaces
text2 = "hello world"
print(text2.replace(" ", ""))  # Output: helloworld


# Example 2: Using maketrans() and translate()
# maketrans() helps to remove many characters at once.

text3 = "h@e#l$l%o123"
remove_chars = "@#$%123"

# Create a table for characters we want to remove
table = str.maketrans('', '', remove_chars)

# Use translate() to remove those characters
print(text3.translate(table))   # Output: hello


# Example 3: Remove all alphabets
text4 = "abc123xyz456"
letters = "abcdefghijklmnopqrstuvwxyz"
table2 = str.maketrans('', '', letters)
print(text4.translate(table2))   # Output: 123456


# Example 4: Remove all numbers
text5 = "abc123xyz456"
numbers = "0123456789"
table3 = str.maketrans('', '', numbers)
print(text5.translate(table3))   # Output: abcxyz
```
