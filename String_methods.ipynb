# 4. Count()

```python
# With the help of count() method, you can count the occurrences of a substring in a string
# Syntax: string.count('value', start, end)

# 'value' = substring or character you want to count
# start   = starting index (optional)
# end     = ending index (optional)

string = "Hello this is python class"

print(string.count('l'))        # counts 'l' in the whole string
print(string.count('l', 0, 5))  # counts 'l' from index 0 to 4
```

# 5. endswith()

```python
# It returns boolean value 
# It checks whether the string is ending with the specified character or not.
# If its ending with the specified Character then it Returns "True"
# If its not ending with the specified Character then it Returns "False"
# You can also specify the range (start , end)

string = "Hello Python World"
print(string.endswith('d')) # it will return True , because its ending with ('d')
print(string.endswith('y')) # it will return False , because its not ending with ('y')
print(string.endswith('y',0,8)) # it will return True, because its ending with ('y'), It will check 'Hello Py' index(0-7) last index is excluded.
```

# 6. startswith()

```python
# It returns boolean value 
# It checks whether the string is starting with the specified character or not.
# If its starting with the specified Character then it Returns "True"
# If its not starting with the specified Character then it Returns "False"
# You can also specify the range (start , end)

string = "Hello Python World"
print(string.startswith('H')) # it will return True , because its starting with ('H')
print(string.startswith('y')) # it will return False , because its not starting with ('y')
print(string.startswith('P',6,)) # it will return True, because its starting with ('P'), It will check 'Python World' index(6-length).
```

# 9. isalpha()

```python
# isalpha check whether the string contains only alphabets (A-Z a-z)
# If yes then it will return True
# If not then it will return False
# These are not alphabets (numbers , space , empty , special characters)  -> False
string1 = "HelloPythonWorld"
string2 = "Hello Python World"
string3 = "Hello@Python@World"
string4 = "HelloPythonWorld123"
print(string1.isalpha()) # -> Output True
print(string2.isalpha()) # -> Output False (space)
print(string3.isalpha()) # ->  Output False (special character)
print(string4.isalpha()) # -> Output False (numbers)
```

# 10. isnumeric()

```python
# isnumeric check whether the string contains only numbers (0-9) , (superscripts , roman numbers)
# If yes then it will return True
# If not then it will return False
# These are not numeric (alphabets , space , empty , special characters, decimal numbers , numbers with plus (+) or minus (−) sign)  -> False

string1 = '123'
string2 = '1 2 3'
string3 = '1@2@3'
string4 = 'Ali123'
string5 = "Ⅻ"
string6 = "²"
string7 = '+123'
string8 = '-123'
string9 = '12.3'
print(string1.isnumeric()) # -> Output True
print(string2.isnumeric()) # -> Output False , contains space
print(string3.isnumeric()) # -> Output False , contains special characters
print(string4.isnumeric()) # -> Output False , contains alphabets
print(string5.isnumeric()) # -> Output True , Roman Number
print(string6.isnumeric()) # -> Output True , superscript
print(string7.isnumeric()) # -> Output False , contains signed value
print(string8.isnumeric()) # -> Output False , contains signed value
print(string9.isnumeric()) # -> Output False , Contains decimal value
```

# 11. isdecimal()

```python
# It returns True only if all the characters in string contain only decimal digits (0-9) or Full width digits "１２３"
# Full-width digits (same as decimal digits but bigger in size) ("１２３")
# If not then it will return False
# These are not decimal digits (alphabets , space , empty , special characters, decimal dot(.), numbers with plus (+) or minus (−) sign ,
# superscripts , roman numbers)  -> False

string1 = '123'
string2 = '12.3'
string3 = '1 2 3'
string4 = 'Ali123'
string5 = "Ⅻ"
string6 = "²"
string7 = '+123'
string8 = '-123'
string9 = "１２３"
print(string1.isdecimal()) # -> Output True , Contains only numbers (0-9)
print(string2.isdecimal()) # -> Output False , Contains (dot , decimal)
print(string3.isdecimal()) # -> Output False , Contains space
print(string4.isdecimal()) # -> Output False , Contains alphabets
print(string5.isdecimal()) # -> Output False , Contains roman numbers
print(string6.isdecimal()) # -> Output False , Contains superscript
print(string7.isdecimal()) # -> Output False , Contains +sign
print(string8.isdecimal()) # -> Output False , Contains -sign
print(string9.isdigit()) # -> Output True , contains Full-width digits
```

# 12. isdigit()

```python
# isdigit Return True only if string contains only :
# 1. Decimal digits (0-9)
# 2. superscripts "²"
# 3. Full-width digits (same as decimal digits but bigger in size) ("１２３")
# These are not digits (alphabets , space , empty , special characters, decimal dot(.), numbers with plus (+) or minus (−) sign ,
# roman numbers)  -> False

string1 = '123'
string2 = '12.3'
string3 = '1 2 3'
string4 = 'Ali123'
string5 = "Ⅻ"
string6 = "²"
string7 = '+123'
string8 = '-123'
string9 = "１２３"

print(string1.isdigit()) # -> Output True  , contains only decimal digits 
print(string2.isdigit()) # -> Output False , contains decimal dot(.)
print(string3.isdigit()) # -> Output False , contains space
print(string4.isdigit()) # -> Output False , contains alphabets
print(string5.isdigit()) # -> Output False , contains roman numbers
print(string6.isdigit()) # -> Output True , contains superscript
print(string7.isdigit()) # -> Output False , contains +sign
print(string8.isdigit()) # -> Output False , contains -sign
print(string9.isdigit()) # -> Output True , contains Full-width digits
```

# 20. find()

```python
# find() searches the substring in a given string and returns the index of its first occurence.
# Difference in find() and index() -> index() raises an error if substring is not found in the string , on the other hand find() return -1.
# syntax string.find(value,start,end)
# value - required
# start - optional
# end - optional

string = "Hello python , this is python class 1 , python batch 2PM"

print(string.find("python"))    # -> Output it will return (6) first occurence.
print(string.find("java"))      # -> Output (-1) substring does not exist
print(string.find("python",7))  # -> Output (23) , it will start checking from index 7
print(string.find("python",24)) # -> Output (40) , it will start checking from index 24
```

# 21. rfind()

```python
# rfind() searches for a substring from right to left and returns the index of its last occurrence.
# syntax string.rfind(value,start,end)
# value - required
# start - optional
# end - optional
string = "Hello python , this is python class 1 , python batch 2PM"

print(string.rfind("python"))    # -> Output it will return (40) last occurrence.
print(string.rfind("java"))      # -> Output (-1) substring does not exist
print(string.rfind("python",0,35))  # -> Output (23) , last occurrence in (0-34) is (23)
print(string.rfind("python",24)) # -> Output (40) , it will start checking from index 24
```

# 23. strip()

```python
# strip() removes characters from both beginning and end of a string.
# string.strip([chars])
# chars → characters to remove (optional)
# If no argument is given → it removes whitespace by default.
# Whitespaces -> space" " , \n 'newline' , \t 'tab' etc
string = "   Hello World   "
string2 = '---Hello World---'

print(string.strip())
print(string2.strip("-"))

#strip() does NOT remove a full word or pattern.

#It removes individual characters.

string3 = "xyxHello Worldxyx"
print(string3.strip("xy"))

# it removes 'x' and 'y' from both ends until it finds a different character.
```

# 26. split()

```python
# split() divides a string into a list of substring using a specified separator.
# syntax string.split(sep=None , maxsplit= -1)
# sep -> Separator (defualt any whitespace)
# maxsplit -> number of split you want to make (default -1) 

string = "Hello Python World"
print(string.split()) # -> Output ['Hello','Python','World'] It will split on every space
print(string.split(" ",1)) # -> Output ['Hello','Python World'] It will split on first space only becuase of maxsplit (1)

string2 = "Hello-Python-World-This-Is-Class1"
print(string2.split("-",3)) # Output ['Hello' , 'Python' , 'World' , 'This-Is-Class1']
```

# 27. rsplit()

```python
# rsplit() divides a string from the right side into a list of substring using a specified separator.
# syntax string.rsplit(sep=None , maxsplit= -1)
# sep -> Separator (defualt any whitespace)
# maxsplit -> number of split you want to make (default -1)

string = "Hello Python World"
print(string.rsplit()) # -> Output ['Hello' , 'Python','World'] It will split on every space
print(string.rsplit(" ",1)) # -> Output ['Hello Python' , 'World'] It will split on first space from right side only becuase of maxsplit (1)

string2 = "Hello-Python-World-This-Is-Class1"
print(string2.rsplit("-",3)) # Output ['Hello-Python-World','This','Is','Class1']

# split() and rsplit() behave differently only when maxsplit is used , otherwise both will gave same result

string3 = "Hello Python World"
print(string3.rsplit()) # -> Output ['Hello' , 'Python','World'] It will split on every space

string4 = "Hello Python World"
print(string4.split()) # -> Output ['Hello','Python','World'] It will split on every space
```

# 30. rjust()

```python
# rjust() shifts the string to right side by padding characters on the left side.
# syntax string.rjust(width,fillchar=" ")
# width = final length of the string
# fillchar = character used for padding (default " ")
# fillchar must be exactly one character long
# if width is smaller than the string length, no trimming happens returns the original string

string = "Hello"
print(string.rjust(10))     # -> Output '     Hello' (default fillchar = ' ')
print(string.rjust(10,'-')) # -> Output '-----Hello'
print(string.rjust(4))      # -> Output 'Hello' (width is smaller than the string length)
```

# 38. partition()

```python
# partition() splits the given string into a tuple of three parts using the given separator.
# 1st part  -> part before the separator
# 2nd part  -> the separator itself
# 3rd part  -> part after the separator
#
# If the separator does not exist in the string,
# it still returns a tuple of 3 elements:
# (whole_string, '', '')

string = "Hello Python World"

print(string.partition("Java"))   # Output -> ('Hello Python World', '', '')
print(string.partition("Python")) # Output -> ('Hello ', 'Python', ' World')
print(string.partition("Hello"))  # Output -> ('', 'Hello', ' Python World')
print(string.partition("World"))  # Output -> ('Hello Python ', 'World', '')
```

# 39. rpartition()

```python
# rpartition() splits the given string into a tuple of three parts using the given separator starting from right side.
# 1st part  -> part before the separator
# 2nd part  -> the separator itself
# 3rd part  -> part after the separator
#
# If the separator does not exist in the string,
# it still returns a tuple of 3 elements:
# ('','',whole_string)

string = "Hello Python World Python World Python"

print(string.rpartition("Java"))   # Output -> ('','','Hello Python World')
print(string.rpartition("Python")) # Output -> ('Hello Python World Python World ', 'Python', '')
print(string.rpartition("Hello"))  # Output -> ('', 'Hello', ' Python World Python World Python')
print(string.rpartition("World"))  # Output -> ('Hello Python World Python ', 'World', ' Python')
```

# 42. maketrans() and translate()

```python
# maketrans() and translate()

# maketrans() creates a translation table that defines how characters should be replaced or removed.
# syntax str.maketrans("characters to be replaced" , "Characters to be replaced with" , "Characters to be removed")

# translate() uses that translation table to replace or remove characters in a string and returns a new modified string.
# syntax string.translate(table)

# Example 1: Replacing characters
string = "apple banana"

table = str.maketrans("ab", "xy")   # a -> x , b -> y
print(string.translate(table))
# Output -> xpple yxnxnx


# Example 2: Removing characters
string2 = "hello world"

table2 = str.maketrans("", "", "aeiou")  # remove vowels
print(string2.translate(table2))
# Output -> hll wrld


# Example 3: Replace and remove together
string3 = "hello world"

table3 = str.maketrans("hw", "HW", "eo")
# h -> H , w -> W , remove e and o
print(string3.translate(table3))
# Output -> Hll Wrld
```

# 45. zfill()

```python
# zfill() pads the string with zero on the left side.
# syntax --> string.zfill(width)
# width --> final length of the string after padding
# if the string starts with (+ or -) , padding will start after the signs.
# If width is less than or equal to the string length, the original string is returned.

string = "50"
print(string.zfill(4))

# Padding will start after (+)
string2 = "+50"
print(string2.zfill(4))

string3 = "Hello World"
print(string3.zfill(20))

string4 = "Hello World" # width is equal to the actual length of the string , it will return the original string
print(string3.zfill(11))
```

# 46. splitlines()

```python
# splitlines() is used to split a string into a list of lines.
# It breaks the string wherever a line boundary occurs (like newline characters).
# \n --> Newline --> Moves the cursor to the next line (moves to nextline in linux , macOs ,modern systems)
# \r\n --> carriage return and newline --> \r moves the cursor to the start of the line , \n moves the cursor to the next line
# \f --> Form feed --> Moves the cursor to new page
# etc

# (Keepends --> ??? (True , False)
# False(defualt) --> newline characters not mentioned in output
# True --> newline characters will be mentioned in ouput) ---->> ????

string = "Hello\fWorld"# \f --> form feed ---> Output ['Hello','World']
print("Hello\fWorld")  # working of \f
print(string.splitlines()) 

string2 = "Hello Python\nWorld" # \n --> newline --->  Output ['Hello Python','World']
print("Hello Python\nWorld") # working of \n
print(string2.splitlines())

string3 = "Hello\r\nWorld"
print("Hello\r\nWorld") # working of \r\n
print(string3.splitlines())
```
