# Q1. Write a program to calculate the electricity bill using only if statement?

```python
Number_of_Units = int(input("Enter number of units : "))

if Number_of_Units <= 100:
    print("No Charges")
elif Number_of_Units > 100 and Number_of_Units <= 200:
    Bill = (Number_of_Units - 100) * 5
    print(f"Bill due is Rs.{Bill}")
else:
    Bill = (Number_of_Units - 200) * 10 + (100 * 5)
    print(f"Bill due is Rs.{Bill}")
```

```python
"Using only if statement"
Units = int(input("Enter Units : "))
if Units >200:
    print("Bill Due :",(Units-200) *10 + (100 *5))
if Units > 100 and Units <= 200:
    print("Bill Due :",(Units - 100)*5)
if Units <= 100:
    print("No Charge")
```

# Q2. Write a program to accept percentage from the user and display the grade.

```python
"""Criteria 
Marks             Grade
> 90               A
> 80 and <= 90     B
>= 60 and <= 80    C
below 60           D"""

Student_percentage = float(input("Enter percentage : "))

if Student_percentage > 90:
    print("Grade A")
elif Student_percentage > 80 and Student_percentage <= 90:
    print("Grade B")
elif Student_percentage >= 60 and Student_percentage <= 80:
    print("Grade C")
else:
    print("Grade D")
```

# Doubt : What if two or more people have the same youngest age , Eg: 18,18,18,25

# Q4. A company decided to give bonus to employee according to following criteria:

```python
"""
Time period of Service      Bonus

More than 10 years           10%

>=6 and <=10                  8%

Less than 6 years             5%
"""

T_O_S = float(input("Enter Time Period Of Service : "))
Salary = float(input("Enter Your Salary : "))

if T_O_S > 10:
    Bonus = Salary * 0.10
    print(f"Net Bonus Amount is : {Bonus}")
elif T_O_S >=6 and T_O_S <= 10:
    Bonus = Salary * 0.08
    print(f"Net Bonus Amount is : {Bonus}")
else:
    Bonus = Salary * 0.05
    print(f"Net Bonus Amount is : {Bonus}")
```

# Q6. Accept the marked price from the user and calculate the Net amount as (Marked Price – Discount) to pay according to following criteria:

```python
"""
Marked Price       Discount
>10000              20%
>7000 and <=10000   15%
<=7000              10%
"""

Price = float(input("Enter marked price : "))

if Price > 10000:
    Discount = Price * 0.20
    Net_Amount = (Price - Discount)
    print(f"Net Amount is {Net_Amount}")
elif Price > 7000 and Price <= 10000:
    Discount = Price * 0.15
    Net_Amount = (Price - Discount)
    print(f"Net Amount is {Net_Amount}")
else:
    Discount = Price * 0.10
    Net_Amount = (Price - Discount)
    print(f"Net Amount is {Net_Amount}")
```

# Q7. Accept the marks of English, Math and Science, Social Studies Subject and display the stream allotted according to following:

```python
"""
All Subjects more than 80 marks         — Science Stream

English >80 and Math, Science above 50  — Commerce Stream

English > 80 and social studies > 80    — Humanities
"""

English = float(input("Enter Marks : "))
Math = float(input("Enter Marks : "))
Science = float(input("Enter Marks : "))
Social_Studies = float(input("Enter Marks : "))

if English > 80 and Math > 80 and Science > 80 and Social_Studies > 80:
    print("Science Stream")
elif English > 80 and Math > 50 and Science > 50:
    print("Commerce Stream")
elif English > 80 and Social_Studies > 80:
    print("Humanities")
else:
    print("Didn't match the requirements")
```

# Q9. Write a program to check whether the last digit of a number (entered by user) is divisible by 3 or not?

```python
# If you use %10 on any number it will give you the last digit. Eg : 2353 % 10 = 3 (It's just a remainder)
```

```python
Number = int(input("Enter Your digit : "))

last_digit = Number % 10
if (last_digit % 3) == 0:
    print(f"yes last digit of {Number} is divisible by 3")
else:
    print(f"No last digit of  {Number} is not divisible by 3")
```
